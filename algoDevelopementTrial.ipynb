```python
lat = df['lat'] 
long = df['long']
z = df['zaxis']
```

```python
l = []
m = []
n = []
l.append(lat[0])
m.append(long[0])
n.append(z[0])
i=1
def fun():
     for i in range(len(lat)):
            if lat[i] in l:
                index = l.index(lat[i])
                if long[i] == m[index]:
                    n[index]=max(z[i],n[index])
                else:
                    l.append(lat[i])
                    m.append(long[i])
                    n.append(z[i])
            else:
                l.append(lat[i])
                m.append(long[i])
                n.append(z[i])
                
```
